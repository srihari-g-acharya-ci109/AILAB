```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 5, 3, 6]
```

```python
def predict(c, m, x):
    return [c+m*xi for xi in x]
```

```python
def mse(actual, predicted):
    n= len(actual)
    s= 0
    for i in range(n):
        s+= (actual[i]-predicted[i])**2
    return s/n
```

```python
def linear_regression(x, y):
    n= len(x)
    x_bar= sum(x)/n
    y_bar= sum(y)/n

    num, den= 0, 0
    for i in range(n):
        num+= (x[i]-x_bar)*(y[i]-y_bar)
        den+= (x[i]-x_bar)**2

    m= num/den
    c= y_bar-m*x_bar

    return c, m
```

```python
def ridge_regression(x, y, lam):
    n= len(x)
    x_bar= sum(x)/n
    y_bar= sum(y)/n

    num, den= 0, 0
    for i in range(n):
        num+= (x[i]-x_bar)*(y[i] - y_bar)
        den+= (x[i]-x_bar)**2

    m= num/(den+lam)
    c= y_bar-m*x_bar

    return c, m
```

```python
def lasso_regression(x, y, lam):
    n= len(x)
    x_bar= sum(x) / n
    y_bar= sum(y) / n

    num, den= 0, 0
    for i in range(n):
        num+= (x[i]-x_bar)*(y[i]-y_bar)
        den+= (x[i]-x_bar)**2

    if abs(num)<=lam:
        m= 0
    else:
        m = (abs(num) - lam) / den
        if num < 0:
            m= -m

    c= y_bar-m*x_bar
    return c, m
```

```python
lam = 1

# Linear Regression
c_lin, m_lin= linear_regression(x, y)
pred_lin= predict(c_lin, m_lin, x)
mse_lin= mse(y, pred_lin)

# Ridge Regression
c_rid, m_rid= ridge_regression(x, y, lam)
pred_rid= predict(c_rid, m_rid, x)
mse_rid= mse(y, pred_rid)

# Lasso Regression
c_las, m_las= lasso_regression(x, y, lam)
pred_las= predict(c_las, m_las, x)
mse_las= mse(y, pred_las)
```

```python
print("Linear Regression")
print("c=", round(c_lin, 4), "m=", round(m_lin, 4))
print("Predicted=", [round(val, 4) for val in pred_lin])
print("MSE=", round(mse_lin, 4))
print("-"*50)

print("Ridge Regression (位=1)")
print("c=", round(c_rid, 4), "m=", round(m_rid, 4))
print("Predicted=", [round(val, 4) for val in pred_rid])
print("MSE=", round(mse_rid, 4))
print("-"*50)

print("Lasso Regression (位=1)")
print("c=", round(c_las, 4), "m=", round(m_las, 4))
print("Predicted=", [round(val, 4) for val in pred_las])
print("MSE=", round(mse_las, 4))
```

```python
errors = {
    "Linear Regression": mse_lin,
    "Ridge Regression": mse_rid,
    "Lasso Regression": mse_las
}
best = min(errors, key=errors.get)

print("Best Model based on MSE= ", best)
```
